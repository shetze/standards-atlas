```python
from sentence_transformers import SentenceTransformer
import torch
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)
```

```python
def generate_embeddings(model_name, texts, batch_size=16):
    """Generate embeddings in batches for a list of texts using a given model."""
    model = SentenceTransformer(model_name).to(device)
    embeddings = []
    
    for i in tqdm(range(0, len(texts), batch_size), desc=f"Encoding with {model_name}"):
        batch = texts[i:i+batch_size]
        batch_embeddings = model.encode(batch, convert_to_numpy=True, show_progress_bar=False)
        embeddings.extend(batch_embeddings)
    
    return embeddings
```

```python
df.fillna("", inplace=True)
```

```python
embedding_models = ['intfloat/multilingual-e5-large', 'Snowflake/snowflake-arctic-embed-l-v2.0', 'sentence-transformers/all-MiniLM-L12-v2']
```

```python
for model_name in embedding_models:
    # Process 'Clause Text'
    df[f'Embedding_{model_name}_Clause'] = generate_embeddings(model_name, df['Clause Text'].tolist())

    # Process 'Human Heading'
    df[f'Embedding_{model_name}_Heading'] = generate_embeddings(model_name, df['Human Heading'].tolist())
```

```python
df.to_csv('data/clauses_w_embeddings_e5_snowflake_minilm.csv')
```
